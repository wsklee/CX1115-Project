# Drop columns

```python
df = df.drop(['status', 'runtime','production_companies','spoken_languages','homepage','budget','original_title',
             'vote_average','vote_count','production_countries'], axis = 1)
```

# Formatting columns
genres

```python
from ast import literal_eval
df['genres'] = df['genres'].apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x,list) else '')
```

```python
df.head(100)
```

```python
df['keywords'] = df['keywords'].apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x,list) else '')
```
